```python
def stock(symbol):
    df = yf.Ticker(symbol)

def rev_growth(symbol):
    stock = yf.Ticker(symbol)
    rev_growth = stock.info['revenueGrowth']
    rev_growth = round(rev_growth * 100, 2)
    print(str(symbol).upper() + ' has a revenue growth of ' + str(rev_growth) + '%')
```
