```python
plt.figure(figsize=(10, 8))

plt.bar('Lowes', low.info['ebitda'], color='blue')
plt.bar('Home Depot', hd.info['ebitda'], color='orange')


plt.xticks(
    rotation=45, 
    horizontalalignment='right',
    fontweight='light',
    fontsize='x-large'  
)

plt.xlabel('Company')
plt.ylabel('EBITDA (in $ billions)')
```

```python
fig, ax = plt.subplots(figsize =(16, 5))

ax.barh('Lowes', low.info['ebitda'] / 1000000000, color='blue')
ax.barh('Home Depot', hd.info['ebitda'] / 1000000000, color='orange')

# Remove axes splines
for s in ['top', 'bottom', 'left', 'right']:
    ax.spines[s].set_visible(False)
# Remove x, y Ticks
ax.xaxis.set_ticks_position('none')
ax.yaxis.set_ticks_position('none')
# Add padding between axes and labels
ax.xaxis.set_tick_params(pad = 5)
ax.yaxis.set_tick_params(pad = 10)
# Add x, y gridlines
ax.grid(b = True, color ='grey',
        linestyle ='-.', linewidth = 1,
        alpha = 0.2)
# Show top values
ax.invert_yaxis()
# Add annotation to bars
for i in ax.patches:
    plt.text(i.get_width()+0.2, i.get_y()+0.5,
             str(round((i.get_width()), 2)),
             fontsize = 10, fontweight ='bold',
             color ='black')

# Add Plot Title
ax.set_title('Lowes (LOW) vs Home Depot (HD) EBITDA (in $ USD billions) FY 2021',
             loc ='left', )

plt.show()
```
